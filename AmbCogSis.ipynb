```python
import sys
sys.path.append('./bin')
import matplotlib.pyplot as plt
%matplotlib inline
import Python_sml_ClientInterface as sml
import numpy as np
import time
import math

k = sml.Kernel.CreateKernelInNewThread()

#raw_input("Conecte o kernel no Soar debugger (atraves do menu Kernel) e pressione uma tecla.")

agent = k.CreateAgent('soar')
answer = agent.LoadProductions('projSisCog_RL.soar')

if answer==False:
    raw_input("Erro ao carregar arquivo SOAR. Carregue manualmente no debugger e entao pressione alguma tecla.")
    
in_Soar = agent.GetInputLink()
position = agent.CreateIntWME(in_Soar, "position", -1)
agent.Commit()
agent.RunSelf(2)   
```

```python
scoreList = []
N = 250# numero de treinos
posAlvo = 6
inicioX = 1 #inicio e fim nao adianta somente alterar aqui, estatico no soar
fimX = 11 
tMovePos = 50./100000 #tempo para mover 1 posicao


ruidoMean = 0
ruidoSigma = 1
noise = np.random.normal(ruidoMean,ruidoSigma)
n_trains_array = range(N)


for j in n_trains_array:
    trialScore = []
    print "Inicio do treino {0}\n".format(j)

    i = inicioX;
    while i < fimX + 1:
        print "Posicao Real: {0}".format(i)
        
        noise = int(math.floor(np.random.normal(ruidoMean,ruidoSigma)))
        
        posRuido = i + noise
        
        if posRuido < inicioX:
            posRuido = inicioX
        if posRuido > fimX:
            posRuido = fimX 
            
        
        print "Posicao com Ruido: {0}".format(posRuido)
        
        agent.Update(position, posRuido)
        agent.Commit()
        antes = time.time()
        agent.RunSelfTilOutput()        
        command = agent.GetCommand(0)
        action_Soar = command.GetParameterValue('action')
        command.AddStatusComplete()

        tempoRespSoar = time.time() - antes;
        posMovidas = int(math.floor(tempoRespSoar/tMovePos));
        
        print "Resposta do Soar durou: {}".format(tempoRespSoar)
        print "O alvo se moveu {} posicoes enquanto o soar respondia\n".format(posMovidas)
        i = i + posMovidas;
        
        if action_Soar == 'fire':
            print "Atirou"
            
            if i == posAlvo:
                score_Soar = 5
            else:
                if i == (posAlvo - 1) or i == (posAlvo + 1):
                    score_Soar = 1
                else:
                    score_Soar = -1                
            score = agent.CreateIntWME(in_Soar, "score", score_Soar)
            agent.Commit()
            agent.RunSelf(1)            
            trialScore.append(score_Soar)
            print "Pontuacao: {0}\n".format(score_Soar)
            agent.DestroyWME(score) 
        else:
            print "Esperou\n"
        i += 1
        #raw_input("continuar")
    print "Fim do treino {0}\n".format(j)
    pontuacaoTrial = np.sum(trialScore)
    print "Pontuacao trial: {}\n".format(pontuacaoTrial)
    scoreList.append(pontuacaoTrial)
    #agent.ExecuteCommandLine('init-soar')
    
agent.DestroyWME(position) 
print agent.ExecuteCommandLine('print --rl')
print scoreList

plt.ylim(-5,8)
plt.ylabel('score')
plt.xlabel('trial')
plt.plot(n_trains_array,scoreList,'x-')
```
