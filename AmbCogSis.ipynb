```python
import sys
sys.path.append('./bin')
import matplotlib.pyplot as plt
%matplotlib inline
import Python_sml_ClientInterface as sml
import numpy as np
import time
import math
import re
```

```python
def simSoar(N = 50, posAlvo = 6, timeRunTrial = 4.0, ruidoSigma = 1, debug = False, tReacao = 0.36, timeResPresentation = 0.013):
    """Executa uma simulacao do SOAR, onde ele aprende a atirar no alvo

    Keyword arguments:
    N -- numero de trials
    posAlvo -- posicao na qual atirando acerto o centro do alvo
    tMovePos -- tempo para mover uma posicao
    ruidoSigma -- sigma da destribuição gaussiana que gera erro na percepcao da posivao do alvo
    """
    k = sml.Kernel.CreateKernelInNewThread()        
    agent = k.CreateAgent('soar')
    answer = agent.LoadProductions('projSisCog_RL.soar')
    
    if answer==False:
        raw_input("Erro ao carregar arquivo SOAR. Carregue manualmente no debugger e entao pressione alguma tecla.")
        
    in_Soar = agent.GetInputLink()
    position = agent.CreateIntWME(in_Soar, "position", -1)
    agent.Commit()
    agent.RunSelf(2)       
    scoreList = []
    tirosList = []
    acertosCentroAlvoList = []
    acertosBordaAlvoList = []
    timeMeanList = []
    timeStdList = []
    pesosFireList = []
    firesTimeList = []
    
    inicioX = 1 #inicio e fim nao adianta somente alterar aqui, estatico no soar
    fimX = 1024        
    ruidoMean = 0    
    noise = np.random.normal(ruidoMean,ruidoSigma)
    n_trains_array = range(N)
    
    
    for j in n_trains_array:
        trialScore = []
        trialPosMovidas = []
        trialTime = []
        trialFiresTime = []
        tirosTrial = 0
        acertosCentroAlvoTrial = 0
        acertosBordaAlvoTrial = 0
        if debug: print "Inicio do treino {0}\n".format(j)
        i = inicioX;        
        
        timeStep = 0.001;
        timeRunTrial = round(timeRunTrial/timeResPresentation) * timeResPresentation;
        speed = fimX/ timeRunTrial;
        tMovePos = round(timeRunTrial/fimX, 3)
        itCounter = 0;
        
        while timeStep < timeRunTrial or itCounter > 1000:
            
            itCounter +=1
            if int(timeStep*1000) % int(timeResPresentation*1000) != 0:
                timeStep = ((int(timeStep*1000) / int(timeResPresentation*1000)) + 1) * timeResPresentation
                
        
            i = (speed * timeStep)
        
            if debug: print "tempo: {0}".format(timeStep)    
            if debug: print "velocidade: {0} pixels/s".format(speed)        
            if debug: print "Posicao Real: {0}".format(round(i))                        
            
            noise = np.random.normal(ruidoMean,ruidoSigma)
            posRuido = int(round(i + noise))
            
            if posRuido < inicioX:
                posRuido = inicioX
            if posRuido > fimX:
                posRuido = fimX                             
            
            if debug: print "Posicao transmitida com Ruido: {0}".format(posRuido)
            
            agent.Update(position, posRuido)
            agent.Commit()
            antes = time.time()
            agent.RunSelfTilOutput()        
            command = agent.GetCommand(0)
            action_Soar = command.GetParameterValue('action')
            
            command.AddStatusComplete()
            tempoRespSoar = time.time() - antes;
            tempoRespSoar = math.ceil(tempoRespSoar * 1000) / 1000.0
            if debug: print "Resposta cognitiva durou: {}".format(tempoRespSoar)
            if action_Soar == 'fire':
                tempoTotalRespSoar = round(tempoRespSoar +  tReacao, 3);
                
                if timeStep + tempoTotalRespSoar > timeRunTrial:
                    break
                
                posMovidas = int(math.floor(tempoTotalRespSoar/tMovePos));
                trialPosMovidas.append(posMovidas);
                trialTime.append(tempoRespSoar);
                i = i + posMovidas;
                
                
                if debug: print "Atirou"
                tirosTrial += 1 
                
                if i <= posAlvo + 23 and i >= posAlvo - 24: #tamanho do bullseye é de 20 pixels
                    score_Soar = 5
                    acertosCentroAlvoTrial += 1 
                else:
                    if i <= (posAlvo + 23 + 35) and i > (posAlvo +23) or  i >= (posAlvo - 24 - 36) and i < (posAlvo - 24): #tamanho das laterais do alvo é de 80 --> alvo total 101 pixels
                        score_Soar = 1
                        acertosBordaAlvoTrial += 1
                    else:
                        score_Soar = -1   
                                
                score = agent.CreateIntWME(in_Soar, "score", score_Soar)
                agent.Commit()
                agent.RunSelf(1)            
                trialScore.append(score_Soar)
                trialFiresTime.append(timeStep + round(tempoTotalRespSoar,3) )
                if debug: print "Pontuacao: {0}\n".format(score_Soar)
                agent.DestroyWME(score) 
            else:
                if debug: print "Esperou\n"      
                tempoTotalRespSoar = tempoRespSoar
                posMovidas = int(math.floor(tempoTotalRespSoar/tMovePos));
                trialPosMovidas.append(posMovidas);
                trialTime.append(tempoRespSoar);            
                #i = i + posMovidas;                
            if debug: print "Resposta do Agente durou: {}".format(tempoTotalRespSoar)
            if debug: print "O alvo se moveu {} posicoes enquanto o agente respondia\n".format(posMovidas)
             
            timeStep += round(tempoTotalRespSoar,3);
            #timeStep += timeResPresentation
            
        if debug: print "Fim do treino {0}\n".format(j)
        pontuacaoTrial = np.sum(trialScore)
        posMovidasTrialMean = np.mean(trialPosMovidas)
        posMovidasTrialStd = np.std(trialPosMovidas)
        timeTrialMean = np.mean(trialTime)
        timeTrialStd = np.std(trialTime)
        
        
        
        if debug: print "Pontuacao trial: {}\n".format(pontuacaoTrial)
        if debug: print "Total de tiros: {}\n".format(tirosTrial)
        if debug: print "Tiros que acertou o centro do alvo: {}\n".format(acertosCentroAlvoTrial)
        if debug: print "Tiros que acertou a borda do alvo: {}\n".format(acertosBordaAlvoTrial)
        if debug: print "Media do tempo total que o agente demorou para atuar: {}\n".format(timeTrialMean)
        if debug: print "Desvio do tempo total que o agente demorou para atuar: {}\n".format(timeTrialStd)
        if debug: print "Media do numero de posicoes que o alvo se moveu ate a atuacao do agente: {}\n".format(posMovidasTrialMean)
        if debug: print "Desvio do numero de posicoes que o alvo se moveu ate a atuacao do agente: {}\n".format(posMovidasTrialStd)
                      
        scoreList.append(pontuacaoTrial)        
        tirosList.append(tirosTrial)        
        acertosCentroAlvoList.append(acertosCentroAlvoTrial)
        acertosBordaAlvoList.append(acertosBordaAlvoTrial)
        timeMeanList.append(timeTrialMean)
        timeStdList.append(timeTrialStd)
        firesTimeList.append(trialFiresTime)    
        rl = agent.ExecuteCommandLine('print --rl')
        todosPesos = rl.split('\n')
        pesosFireTrial = []
        for peso in todosPesos:    
            if re.search('fire',peso) != None:
                pesosFireTrial.append(peso.split(' ')[3])
        #inverto a lista p/ ficar na ordem certa 
        pesosFireList.append(pesosFireTrial[::-1])
            
    agent.DestroyWME(position) 
    rl = agent.ExecuteCommandLine('print --rl')
    
    k.Shutdown()
    
    return [scoreList, tirosList, acertosCentroAlvoList, acertosBordaAlvoList, timeMeanList, timeStdList, pesosFireList, firesTimeList]
```

##Simulação Parametros realisticos

```python
qtdTrials = 100

tempoAlvoMoverAteFim = 8.0 #em s

sigmaDoRuido = 0.005
#qto menor, mais concentrada a distribuicao, i.e. menos ruido

tAtuacao = 0.36 #em s, tempo da resposta neuromuscular 

periodoDaApresentacaoEstimulo = 0.013

[score, tiros, acertosCentroAlvo, acertosBordaAlvo, meanTimeResp, stdTimeResp, pesosFire, firesTime] = simSoar(qtdTrials, 512, tempoAlvoMoverAteFim, sigmaDoRuido, False, tAtuacao, periodoDaApresentacaoEstimulo)        
```
